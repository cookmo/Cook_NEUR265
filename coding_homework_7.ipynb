```python
# Importing modules (#2)
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing data (#3)
acc_hits = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/acc_hits_hw5.csv', delimiter = ',')
acc_fa = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/acc_fa_hw5.csv', delimiter = ',')
```

```python
# Printing shape of data sets (#4)
print(acc_hits.shape)
print(acc_fa.shape)
```

```python
# Creating a time variable (#6)
time = np.linspace(0, 4, 8001)
```

```python
# Determining the sampling rate of data (#7)
srate = 8001/4
print(srate)
```

```python
# Plotting first trials in subplots (#8)
fig, ax = plt.subplots(1,2)

ax[0].plot(time, acc_hits[0])
ax[1].plot(time, acc_fa[0])

ax[0].set_xlabel('Time (s)')
ax[0].set_ylabel('Voltage (mV)')

ax[1].set_xlabel('Time (s)')
ax[1].set_ylabel('Voltage (mV)')

plt.tight_layout()
```

9. Looking at the graph, it appears that the dominant oscillation in the LFP may be about ~1-2 Hz for acc_hits data and about ~2–3 Hz for the acc_fa data, meaning that the two differ.

```python
# Calculating and plotting the ERP for acc_hits (#10)
plt.plot(time, acc_hits.mean(0))

plt.vlines(2, -50, 50, color = 'black')
plt.hlines(0, 0, 4, color = 'black')

plt.xlabel('Time (s)')
plt.ylabel('Voltage (mV)')
```

11. After the screen touch takes place, the activity of the LFP becomes more consistent and with larger waves. There appears to be evidence of consistent activity in the LFP surrounding screen touch, as this activity follows right after the screen touch takes place.

```python
# Calculating and plotting the ERP for acc_fa (#12)
plt.plot(time, acc_fa.mean(0))

plt.vlines(2, -10, 10, color = 'black')
plt.hlines(0, 0, 4, color = 'black')

plt.xlabel('Time (s)')
plt.ylabel('Voltage (mV)')
```

13. The magnitudes of change in voltage for the false alarm data is less than that of the changes in the hits data. The response after the stimulus also does not differ much from what it follows. Overall, the response to the correct hits creates stronger LFP activity.
