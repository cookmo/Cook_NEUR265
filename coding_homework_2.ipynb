```python
# Importing numpy (#2)
import numpy as np
import matplotlib.pyplot as plt
import statistics as st
```

```python
# Importing the data as numpy arrays (#3)
url1 = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_sst.csv'
url2 = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_pvalb.csv'

v_rest1 = np.loadtxt(url1)
v_rest2 = np.loadtxt(url2)
```

```python
# Concatenating the data sets into a matrix and checking the shape (#4)
v_rest = np.vstack((v_rest1, v_rest2))

print(v_rest.shape)
```

```python
# Printing the size of the matrix (#5)
print(v_rest.size)
```

```python
# Printing the first row of the matrix (#6)
print(v_rest[0])
```

```python
# Calculate the sum of the third column of the matrix (#7)
v_rest[:, 2].sum()
```

```python
# Printing the second-to-last value in the second row (#8)
v_rest[1, -2]
```

```python
# Creating a histogram of the first row values of the matrix (#9)
plt.hist(v_rest[0])
plt.xlabel('Voltage (mV)')
plt.ylabel('Frequency')
plt.title('SST Resting Memrane Potential Frequency')
```

```python
# Creating new variables for the means of rows (#10)
v_rest_mean_sst = st.mean(v_rest[0])
v_rest_mean_pvalb = st.mean(v_rest[1])

print(v_rest_mean_sst)
print(v_rest_mean_pvalb)
```

The mean resting potential for SST neurons is -67.73 mV, while the mean resting potential for Pvalb neurons is -69.98 mV. The SST mean is larger than that of the Pvalb mean.

```python
# Importing "Scipy" (#12)
import scipy.stats as stats
```

```python
# Performing a t-test on the two rows (#13)
stats.ttest_ind(v_rest[0], v_rest[1])
```

The results of the t-test are statistically significant. The test resulted in a p-value of 7.16x10^-9, which is much less than the "statistically significant" value of 0.05. So, because p < 0.05, it is likely that the means are significantly different.
